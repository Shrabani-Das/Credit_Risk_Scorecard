### 🎯 Objective

Develop an industry-style feature mart for Probability of Default (PD) modeling by integrating applicant, bureau, previous credit history, and repayment behavior data into a unified customer-level analytical dataset.

The engineered feature set is designed to support scorecard development, WOE binning, variable selection, model training, and regulatory-style credit risk assessment.

---

### 📋 Feature Engineering Summary

| Category | Description |
|-----------|-------------|
| 👤 Applicant Profile | Demographics, employment stability, household characteristics and customer profile indicators |
| 💰 Financial Capacity | Income, credit burden, affordability and leverage ratios |
| 📊 External Risk Signals | External bureau scores, enquiry behaviour and social risk indicators |
| 🏦 Bureau Credit History | Tradelines, exposure, utilization, delinquency, vintage and product-level bureau performance |
| 📑 Previous Applications | Historical application behaviour, approval patterns and prior credit demand |
| 🔄 Repayment Behaviour | Delinquency patterns, DPD trends and installment repayment characteristics |
| 💳 Credit Product Behaviour | Product-level exposure, utilization and portfolio concentration measures |
| 🎯 Target Variable | Customer reaches 60+ DPD within a 12-month performance window |

### Capacity Features

```python
df_temp['credit_income_ratio'] = (
    df_temp['AMT_CREDIT'] /
    df_temp['AMT_INCOME_TOTAL']
)

df_temp['annuity_income_ratio'] = (
    df_temp['AMT_ANNUITY'] /
    df_temp['AMT_INCOME_TOTAL']
)

df_temp['goods_income_ratio'] = (
    df_temp['AMT_GOODS_PRICE'] /
    df_temp['AMT_INCOME_TOTAL']
)

df_temp['credit_goods_ratio'] = (
    df_temp['AMT_CREDIT'] /
    df_temp['AMT_GOODS_PRICE']
)

application_df = application_df.merge(
    df_temp[['SK_ID_CURR','credit_income_ratio','annuity_income_ratio','goods_income_ratio','credit_goods_ratio']], on = 'SK_ID_CURR', how = 'left'
)

application_df.shape
```

### Demographic Features

```python
df_temp['age_years'] = abs(df_temp['DAYS_BIRTH']) / 365

df_temp['age_band'] = pd.cut(
    df_temp['age_years'],
    bins=[18,25,35,45,55,100]
)

df_temp = df_temp.rename(columns={'CNT_FAM_MEMBERS': 'count_family_members'})

df_temp['per_capita_income'] = (
    df_temp['AMT_INCOME_TOTAL'] /
    df_temp['count_family_members']
)


df_temp['children_ratio'] = (
    df_temp['CNT_CHILDREN'] /
    df_temp['count_family_members']
)

application_df = application_df.merge(
    df_temp[['SK_ID_CURR','age_years','count_family_members',
             'per_capita_income','children_ratio']], on = 'SK_ID_CURR', how = 'left'
)

application_df.shape
```

### Stability Features

```python
df_temp['employment_years'] = (
    abs(df_temp['DAYS_EMPLOYED']) /
    365
)

df_temp['employment_age_ratio'] = (
    abs(df_temp['DAYS_EMPLOYED']) /
    abs(df_temp['DAYS_BIRTH'])
)

df_temp['registration_years'] = (
    abs(df_temp['DAYS_REGISTRATION']) /
    365
)

df_temp['id_publish_years'] = (
    abs(df_temp['DAYS_ID_PUBLISH']) /
    365
)

df_temp['owns_house'] = (
    df_temp['FLAG_OWN_REALTY'] == 'Y'
).astype(int)

df_temp['owns_car'] = (
    df_temp['FLAG_OWN_CAR'] == 'Y'
).astype(int)

df_temp = df_temp.rename(columns={'OWN_CAR_AGE': 'own_car_age'})


application_df = application_df.merge(
    df_temp[['SK_ID_CURR','employment_years','employment_age_ratio',
             'registration_years','id_publish_years','owns_house','owns_car','own_car_age']], on = 'SK_ID_CURR', how = 'left'
)

application_df.shape
```

```python
# 1. Create temporary flags for active and closed loans before grouping
bureau["is_active"] = (bureau["CREDIT_ACTIVE"] == "Active").astype(int)
bureau["is_closed"] = (bureau["CREDIT_ACTIVE"] == "Closed").astype(int)

# 2. Run a SINGLE unified aggregation step
bureau_features = (
    bureau.groupby("SK_ID_CURR")
    .agg(
        total_tradelines=("SK_ID_BUREAU", "count"),
        total_active_tradelines=("is_active", "sum"),
        total_closed_tradelines=("is_closed", "sum"),
        total_credit_exposure=("AMT_CREDIT_SUM", "sum"),
        total_debt=("AMT_CREDIT_SUM_DEBT", "sum"),
        total_overdue=("AMT_CREDIT_SUM_OVERDUE", "sum"),
        max_overdue=("AMT_CREDIT_SUM_OVERDUE", "max"),
        avg_credit_overdue_days=("CREDIT_DAY_OVERDUE", "mean"),
        max_credit_overdue_days=("CREDIT_DAY_OVERDUE", "max"),
        credit_prolongation_count=("CNT_CREDIT_PROLONG", "sum"),
        avg_days_credit=("DAYS_CREDIT", "mean"),
    )
    .reset_index()
)

# 3. Perform ratio calculations and flag assignments on the aggregated dataset
bureau_features["active_pct_tradelines"] = np.where(
    bureau_features["total_tradelines"] > 0,
    bureau_features["total_active_tradelines"]
    / bureau_features["total_tradelines"],
    0,
)

bureau_features["debt_to_utilization_ratio"] = np.where(
    bureau_features["total_credit_exposure"] > 0,
    bureau_features["total_debt"] / bureau_features["total_credit_exposure"],
    0,
)

bureau_features["avg_debt_per_tradeline"] = np.where(
    bureau_features["total_tradelines"] > 0,
    bureau_features["total_debt"] / bureau_features["total_tradelines"],
    0,
)

bureau_features["ever_delinquent_flag"] = (
    bureau_features["max_credit_overdue_days"] > 0
).astype(int)

bureau_features["avg_credit_age_years"] = (
    abs(bureau_features["avg_days_credit"]) / 365
)

# Drop the helper intermediate column used for age math
bureau_features.drop(columns=["avg_days_credit"], inplace=True)
bureau_features.shape
```

```python
credit_type_map = {

    # Personal Loans
    'Consumer credit':'PL',
    'Cash loan (non-earmarked)':'PL',
    'Microloan':'PL',

    # Credit Cards
    'Credit card':'CC',

    # Home Loans
    'Mortgage':'HL',
    'Real estate loan':'HL',

    # Auto Loans
    'Car loan':'AL',

    # Business Loans
    'Loan for business development':'BL',
    'Loan for working capital replenishment':'BL',

    # Investment / Secured
    'Loan for purchase of shares (margin lending)':'INV',
    'Loan for the purchase of equipment':'INV',

    # Others
    'Another type of loan':'OTH',
    'Unknown type of loan':'OTH',
    'Mobile operator loan':'OTH',
    'Interbank credit':'OTH'
}

bureau['product_group'] = (
    bureau['CREDIT_TYPE']
    .map(credit_type_map)
)

bureau['tradeline_utilization'] = np.where(
    bureau['AMT_CREDIT_SUM'] > 0,
    bureau['AMT_CREDIT_SUM_DEBT'] /
    bureau['AMT_CREDIT_SUM'],
    np.nan
)

product_features = bureau.groupby(
    ['SK_ID_CURR','product_group']
).agg(

    num_accounts=('SK_ID_BUREAU','count'),

    total_exposure=('AMT_CREDIT_SUM','sum'),

    total_debt=('AMT_CREDIT_SUM_DEBT','sum'),

    total_overdue=('AMT_CREDIT_SUM_OVERDUE','sum'),

    avg_utilization=('tradeline_utilization','mean')

).reset_index()

product_features['utilization'] = np.where(
    product_features['total_exposure'] > 0,
    product_features['total_debt'] /
    product_features['total_exposure'],
    np.nan
)

bureau_prod = product_features.pivot(
    index='SK_ID_CURR',
    columns='product_group',
    values=[
        'num_accounts',
        'total_exposure',
        'total_debt',
        'total_overdue',
        'utilization'
    ]
)

bureau_prod.columns = [

    f"{metric}_{prod}"

    for metric, prod

    in bureau_prod.columns

]

bureau_prod = bureau_prod.reset_index()


active_accounts = (

    bureau[
        bureau['CREDIT_ACTIVE']=='Active'
    ]

    .groupby(
        ['SK_ID_CURR','product_group']
    )

    .size()

    .unstack(fill_value=0)

)

active_accounts.columns = [

    f'active_{col.lower()}'

    for col

    in active_accounts.columns

]

active_accounts = active_accounts.reset_index()


bureau_prod = bureau_prod.merge(

    active_accounts,

    on='SK_ID_CURR',

    how='left'

)

bureau_prod = bureau_prod.fillna(0)

bureau_prod.shape
```

```python
bureau_features = bureau_features.merge(bureau_prod, on='SK_ID_CURR', how='left')
bureau_features.shape
```

```python
# 1. Map status quickly using a vectorized map
status_map = {
    "X": -999,
    "C": 0,
    "0": 0,
    "1": 30,
    "2": 60,
    "3": 90,
    "4": 120,
    "5": 150,
}
bureau_balance["dpd"] = bureau_balance["STATUS"].map(status_map)

# 2. Vectorize the condition check flags BEFORE grouping (No lambdas!)
bureau_balance["is_30"] = (bureau_balance["dpd"] >= 30).astype(int)
bureau_balance["is_60"] = (bureau_balance["dpd"] >= 60).astype(int)
bureau_balance["is_90"] = (bureau_balance["dpd"] >= 90).astype(int)
bureau_balance["is_120"] = (bureau_balance["dpd"] >= 120).astype(int)
bureau_balance["is_150"] = (bureau_balance["dpd"] >= 150).astype(int)

# 3. Create time-window masks using numpy where
for m in [1, 3, 6, 12]:
    mask = bureau_balance["MONTHS_BALANCE"] >= -m
    bureau_balance[f"dpd_l{m}m"] = np.where(mask, bureau_balance["dpd"], np.nan)
    bureau_balance[f"is_30_l{m}m"] = np.where(mask, bureau_balance["is_30"], 0)
    bureau_balance[f"is_60_l{m}m"] = np.where(mask, bureau_balance["is_60"], 0)
    bureau_balance[f"is_90_l{m}m"] = np.where(mask, bureau_balance["is_90"], 0)
    bureau_balance[f"is_120_l{m}m"] = np.where(mask, bureau_balance["is_120"], 0)
    bureau_balance[f"is_150_l{m}m"] = np.where(mask, bureau_balance["is_150"], 0)

# 4. Execute a SINGLE unified group operation with optimized Cython aggregations ('max' and 'sum')
dpd_features = (
    bureau_balance.groupby("SK_ID_BUREAU")
    .agg(
        # 1 Month Window
        max_dpd_l1m=("dpd_l1m", "max"),
        cnt_30dpd_l1m=("is_30_l1m", "sum"),
        cnt_60dpd_l1m=("is_60_l1m", "sum"),
        cnt_90dpd_l1m=("is_90_l1m", "sum"),
        cnt_120dpd_l1m=("is_120_l1m", "sum"),
        cnt_150dpd_l1m=("is_150_l1m", "sum"),
        # 3 Month Window
        max_dpd_l3m=("dpd_l3m", "max"),
        cnt_30dpd_l3m=("is_30_l3m", "sum"),
        cnt_60dpd_l3m=("is_60_l3m", "sum"),
        cnt_90dpd_l3m=("is_90_l3m", "sum"),
        cnt_120dpd_l3m=("is_120_l3m", "sum"),
        cnt_150dpd_l3m=("is_150_l3m", "sum"),
        # 6 Month Window
        max_dpd_l6m=("dpd_l6m", "max"),
        cnt_30dpd_l6m=("is_30_l6m", "sum"),
        cnt_60dpd_l6m=("is_60_l6m", "sum"),
        cnt_90dpd_l6m=("is_90_l6m", "sum"),
        cnt_120dpd_l6m=("is_120_l6m", "sum"),
        cnt_150dpd_l6m=("is_150_l6m", "sum"),
        # 12 Month Window
        max_dpd_l12m=("dpd_l12m", "max"),
        cnt_30dpd_l12m=("is_30_l12m", "sum"),
        cnt_60dpd_l12m=("is_60_l12m", "sum"),
        cnt_90dpd_l12m=("is_90_l12m", "sum"),
        cnt_120dpd_l12m=("is_120_l12m", "sum"),
        cnt_150dpd_l12m=("is_150_l12m", "sum"),
    )
    .reset_index()
)

# Optional clean up: drop temporary columns added to bureau_balance to free RAM
drop_cols = [c for c in bureau_balance.columns if "is_" in c or "m" in c]
bureau_balance.drop(columns=drop_cols, inplace=True, errors="ignore")
```

```python
bureau_enriched = bureau.merge(
    dpd_features,
    on='SK_ID_BUREAU',
    how='left'
)

bureau_enriched.shape
```

```python
windows = ['l1m', 'l3m', 'l6m', 'l12m']

agg_dict = {}

# Maximum DPD
for w in windows:
    agg_dict[f'max_dpd_{w}'] = (f'max_dpd_{w}', 'max')

# Count DPDs
for w in windows:
    for dpd in [30, 60, 90]:
        agg_dict[f'total_{dpd}dpd_{w}'] = (
            f'cnt_{dpd}dpd_{w}',
            'sum'
        )

customer_dpd = (
    bureau_enriched
    .groupby('SK_ID_CURR')
    .agg(**agg_dict)
    .reset_index()
)

customer_dpd.shape
```

```python
major_products = (
    bureau_enriched['product_group']
    .dropna()
    .unique()
    .tolist()
)

windows = ['l1m', 'l3m', 'l6m', 'l12m']

prod_frames = []

for prod in major_products:

    agg_dict = {}

    # Max DPD Features
    for w in windows:
        agg_dict[f'max_dpd_{w}'] = (
            f'max_dpd_{w}',
            'max'
        )

    # Count DPD Features
    for w in windows:
        for dpd in [30, 60, 90]:
            agg_dict[f'cnt_{dpd}dpd_{w}'] = (
                f'cnt_{dpd}dpd_{w}',
                'sum'
            )

    temp = (
        bureau_enriched[
            bureau_enriched['product_group'] == prod
        ]
        .groupby('SK_ID_CURR')
        .agg(**agg_dict)
        .add_prefix(f'{prod}_')
        .reset_index()
    )

    prod_frames.append(temp)

# Combine all product-level feature tables
product_dpd_features = prod_frames[0]

for df_temp in prod_frames[1:]:
    product_dpd_features = product_dpd_features.merge(
        df_temp,
        on='SK_ID_CURR',
        how='outer'
    )

product_dpd_features = product_dpd_features.fillna(0)

# Merge with existing bureau features
bureau_features_final = (
    bureau_features
    .merge(
        product_dpd_features,
        on='SK_ID_CURR',
        how='left'
    ).merge(
        customer_dpd,
        on='SK_ID_CURR',
        how='left'
    )
    .fillna(0)
)

print("Final Bureau Features Shape :", bureau_features_final.shape)

bureau_features_final.head(2)
```

```python
a = prev.SK_ID_CURR.unique().tolist()
b = prev.SK_ID_PREV.unique().tolist()

c = set(a).intersection(set(b))
len(c)
```

```python
prev_feat = prev.groupby(
    'SK_ID_CURR'
).agg(
    num_prev_apps=('SK_ID_PREV','count')
)

prev['is_approved'] = (
    prev['NAME_CONTRACT_STATUS']
    == 'Approved'
).astype(int)

prev['is_refused'] = (
    prev['NAME_CONTRACT_STATUS']
    == 'Refused'
).astype(int)

prev['is_cancelled'] = (
    prev['NAME_CONTRACT_STATUS']
    == 'Canceled'
).astype(int)

approval_features = prev.groupby(
    'SK_ID_CURR'
).agg(
    num_approved=('is_approved','sum'),
    num_refused=('is_refused','sum'),
    num_cancelled=('is_cancelled','sum')
)

approval_features['approval_rate'] = (
    approval_features['num_approved']
    /
    approval_features.sum(axis=1)
)

approval_features['rejection_rate'] = (
    approval_features['num_refused']
    /
    approval_features.sum(axis=1)
)

amount_features = prev.groupby(
    'SK_ID_CURR'
).agg(
    avg_application_amt=('AMT_APPLICATION','mean'),
    max_application_amt=('AMT_APPLICATION','max'),

    avg_credit_amt=('AMT_CREDIT','mean'),
    max_credit_amt=('AMT_CREDIT','max')
)

prev['credit_grant_ratio'] = (
    prev['AMT_CREDIT']
    /
    prev['AMT_APPLICATION']
)

grant_features = prev.groupby(
    'SK_ID_CURR'
).agg(
    avg_credit_grant_ratio=(
        'credit_grant_ratio',
        'mean'
    ),
    min_credit_grant_ratio=(
        'credit_grant_ratio',
        'min'
    )
)

term_features = prev.groupby(
    'SK_ID_CURR'
).agg(
    avg_prev_term=('CNT_PAYMENT','mean'),
    max_prev_term=('CNT_PAYMENT','max')
)

time_features = prev.groupby(
    'SK_ID_CURR'
).agg(
    days_since_last_application=(
        'DAYS_DECISION',
        'max'
    )
)


insurance_features = prev.groupby(
    'SK_ID_CURR'
).agg(
    insurance_rate=(
        'NFLAG_INSURED_ON_APPROVAL',
        'mean'
    )
)

from functools import reduce

prev_feature_list = [

    prev_feat,
    approval_features,
    amount_features,
    grant_features,
    term_features,
    time_features,
    insurance_features

]

previous_application_features = reduce(
    lambda left,right:
    left.merge(
        right,
        left_index=True,
        right_index=True,
        how='outer'
    ),
    prev_feature_list
).reset_index()

previous_application_features.shape
```

### Target Creation : 60+ DPD in 12M Performance Window

```python
# Define the path to your file
file_path = "../Dataset/raw/installments_payments.csv"

# Columns we actually need to build the target
use_cols = ['SK_ID_CURR', 'DAYS_INSTALMENT', 'DAYS_ENTRY_PAYMENT']

# Dictionary to hold aggregated results from each chunk
chunk_targets = []

# Define chunk size (1 million rows at a time uses very little RAM)
chunk_size = 1_000_000

print("Processing installments_payments.csv in chunks...")

for chunk in pd.read_csv(file_path, usecols=use_cols, chunksize=chunk_size):
    # 1. Filter immediately for the 12-month Performance Window
    # (DAYS_INSTALMENT between -365 and 0)
    chunk_filtered = chunk[(chunk['DAYS_INSTALMENT'] >= -365) & (chunk['DAYS_INSTALMENT'] <= 0)].copy()
    
    if chunk_filtered.empty:
        continue
        
    # 2. Calculate DPD (Clip at lower bound 0)
    chunk_filtered['DPD'] = chunk_filtered['DAYS_ENTRY_PAYMENT'] - chunk_filtered['DAYS_INSTALMENT']
    chunk_filtered['DPD'] = chunk_filtered['DPD'].clip(lower=0)
    
    # 3. Create the 60+ DPD Flag
    chunk_filtered['bad_60_flag'] = (chunk_filtered['DPD'] >= 60).astype(int)
    
    # 4. Preliminary aggregation at chunk level to save space
    chunk_agg = chunk_filtered.groupby('SK_ID_CURR')['bad_60_flag'].max().reset_index()
    
    chunk_targets.append(chunk_agg)

print("Chunks processed. Combining and performing final aggregation...")

# 5. Concatenate all chunk results
combined_df = pd.concat(chunk_targets, ignore_index=True)

# 6. Final aggregation (in case a customer appeared across different chunks)
target_df = combined_df.groupby('SK_ID_CURR').agg(TARGET=('bad_60_flag', 'max')).reset_index()

print(f"Target dataframe created successfully! Shape: {target_df.shape}")
print(f"Bad Rate: {target_df['TARGET'].mean() * 100:.2f}%")
```

### Target Definition (Revisited)

Roll-rate analysis identified 60+ DPD as a meaningful deterioration threshold with significant migration into severe delinquency buckets.

Vintage analysis was performed to evaluate portfolio performance stability and support the conceptual selection of a 12-month performance horizon.

Due to limitations in reconstructing a precise account-level forward-looking performance window from the publicly available Home Credit dataset, the final modeling target was defined as:

**TARGET = 1**
- Customer experienced at least one 60+ DPD event in observed repayment history.

**TARGET = 0**
- Customer never experienced a 60+ DPD event.

The resulting bad rate of approximately 2.9% is consistent with severe delinquency-based Probability of Default (PD) modeling.
